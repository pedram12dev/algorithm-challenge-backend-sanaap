# Checking Four Consecutive 1s (Circular Binary String)

## Description
Given a 7-digit binary string such as `1010111`, determine whether there are four consecutive `1`s.

The string is considered circular, meaning the end connects to the beginning.

Example:
`1010111`

There are `111` at the end and `1` at the beginning.
Since the string is circular, they form `1111`.

Output:
True

Time Complexity: O(n)


```python
def has_four_consecutive_ones(binary_string: str) -> bool:
    doubled = binary_string + binary_string
    return "1111" in doubled


if __name__ == "__main__":
    test_cases = [
        "1010111",
        "1111000",
        "0000000",
        "1010101",
        "1100111",
        "0001000",
        "1110011",
    ]

    for input_str in test_cases:
        result = has_four_consecutive_ones(input_str)
        doubled = input_str + input_str

        print(f"input  = \"{input_str}\"")
        print(f"doubled = \"{doubled}\"")
        print(f"output  = {result}")
        print("-" * 20)
```
