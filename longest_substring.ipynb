# Longest Substring Without Repeating Characters

## Description
Given a string such as `ABCABCFKAB`, find the longest substring consisting of consecutive characters with no repeated characters.

Example:
- `ABC` → length 3  
- `ABCFK` → length 5  
- `AB` → length 2  

The longest valid substring is `ABCFK`, so the output should be:

5

Time Complexity: O(n)

```python
def length_longest_substring(s: str) -> int: 
    start = 0 
    max_size = 0 
    char_index = {}

    for i in range(len(s)):
    
        if s[i] in char_index:
            
            start = max(start, char_index[s[i]] + 1)
        
        char_index[s[i]] = i
        
        max_size = max(max_size, i - start + 1)
    
    return max_size


if __name__ == "__main__":
    test_cases_1 = [
        "ABCABCFKAB",
        "ABCABCBB",
        "BBBBB",
        "PWWKEW",
        "",
        "ABCDEF",
        "ABRKAABR"
    ]
    for input_str in test_cases_1:
        length = length_longest_substring(input_str)
        print(f"input = \"{input_str}\"")
        print(f"length = {length}")
```
